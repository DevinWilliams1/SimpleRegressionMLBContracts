```python
# Getting Data
import pandas as pd
```

```python
# Get Excel
df = pd.read_excel('Player_Contracts.xlsx', header=None)
```

```python
# Split Data as it is originally in one column
split_df = df[0].astype(str).str.split(",", expand=True)
split_df
```

```python
# Change column title row
split_df.columns = split_df.iloc[0].values
```

```python
#Drop old column row that was below real columns
split_df = split_df.drop(0, axis = 0)
```

```python
split_df
```

```python
# Drop unneeded columns
columns_drop = ['Rk','Summary','Name-additional']
```

```python
# Drop columns
clean_df = split_df.drop(columns_drop, axis = 1)
clean_df
```

```python
# Convert Last Year to numeric
clean_df['Last Year'] = pd.to_numeric(clean_df['Last Year'], errors='coerce')
```

```python
# Boolean mask where it checks every row, and assigns a values true/false based on 2025. If found to be true, keeps in new data frame
clean_df = clean_df[clean_df['Last Year'] >= 2025]
clean_df
```

```python
# Drop empty WAR Values
clean_df = clean_df[clean_df['WAR'] != '']
clean_df
```

```python
# Reset Index numbers
clean_df = clean_df.reset_index(drop=True)
clean_df
```

```python
# Drop Strasburg due to retirement
clean_df = clean_df.drop(18, axis = 0)
clean_df
```

```python
# Drop Strasburg due to retirement
clean_df = clean_df.drop(27, axis = 0)
clean_df
```

```python
# All players that played up until last year, have not retired, and not suspended (Franco)
clean_df = clean_df.reset_index(drop=True)
clean_df
```

```python
# Player lookup 
from pybaseball import playerid_lookup
```

```python
clean_df = clean_df.drop(6,axis=0)
clean_df
```

```python
clean_df['First Year'] = pd.to_numeric(clean_df['First Year'], errors='coerce')
```

```python
clean_df
```

```python
clean_df['Total Value'] = clean_df['Total Value'].str.replace('$', '').astype(int)
```

```python
clean_df['Avg Annual Value'] = clean_df['Avg Annual Value'].str.replace('$', '').astype(int)
```

```python
clean_df['WAR/yr'] = clean_df['WAR/yr'].astype(float)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
y = clean_df[['Avg Annual Value']]
x = clean_df[['WAR/yr']]
```

```python
reg = LinearRegression().fit(x,y)
```

```python
print("Coefficient (slope):", reg.coef_[0])
print("Intercept:", reg.intercept_)
print("R-squared:", reg.score(x, y))
```

```python
y1 = clean_df[['Avg Annual Value']]
x1 = clean_df[['First Year']]
```

```python
reg1 = LinearRegression().fit(x1,y1)
```

```python
print("Coefficient (slope):", reg1.coef_[0])
print("Intercept:", reg1.intercept_)
print("R-squared:", reg1.score(x1, y1))
```

```python
y2 = clean_df['Avg Annual Value']
x2 = clean_df[['First Year', 'WAR/yr']]
```

```python
reg2 = LinearRegression().fit(x2,y2)
```

```python
print("Coefficient (slope):", reg2.coef_[0])
print("Intercept:", reg2.intercept_)
print("R-squared:", reg2.score(x2, y2))
```

```python
#Plot multiple regression
import matplotlib.pyplot as plt
```

```python
# Generate predictions for the second model
y2_pred = reg2.predict(x2)

plt.figure(figsize=(10, 6))
plt.scatter(y2, y2_pred, alpha=0.7)
plt.plot([y2.min(), y2.max()], [y2.min(), y2.max()], 'r--', linewidth=2, label='Perfect prediction')

plt.xlabel('Actual Avg Annual Value')
plt.ylabel('Predicted Avg Annual Value')
plt.title(f'Multiple Regression: Predicted vs Actual\nR² = {reg2.score(x2, y2):.4f}')
plt.legend()
plt.grid(True, alpha=0.5)
plt.show()
```
